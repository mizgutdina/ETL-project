```python

import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

# ****STARBUCKS FILE***

```python

# reviewing Starbucks file

csv_file = "resources/Starbucks_data.csv"
starbucks_df = pd.read_csv(csv_file)
starbucks_df
```

```python
# filter location "Country" US 
us_starbucks_df = starbucks_df.loc[starbucks_df["Country"] == "US"]
```

```python
#Drop NaN values 

us_starbucks_df['Postcode'] = us_starbucks_df['Postcode'].dropna()

us_starbucks_df['Postcode'] 
```

```python
# need to retrieve only 5 digit zipcodes from 9-digit Postcodes

us_starbucks_df['five_digit_zip'] = np.where(us_starbucks_df['Postcode'].str.len() == 9, 
                                     us_starbucks_df['Postcode'].str[:5],
                                     us_starbucks_df['Postcode'])
print(us_starbucks_df)
```

```python
# need to retrieve only 5 digit zipcodes from 8-digit Postcodes

us_starbucks_df['five_digit_zip'] = np.where(us_starbucks_df['five_digit_zip'].str.len() == 8, 
                                     us_starbucks_df['five_digit_zip'].str[:4],
                                     us_starbucks_df['five_digit_zip'])
print(us_starbucks_df)
```

```python
#Add  0 to zip code to 4-digit Postcode
us_starbucks_df['five_digit_zip'] = us_starbucks_df['five_digit_zip'].str.zfill(5)
us_starbucks_df['five_digit_zip']
```

```python
#Pick the relevant column names
clean_us_starbucks_df = us_starbucks_df[['Store Number','Store Name','Street Address','City','State/Province','Country', 'five_digit_zip', 'Phone Number',
                    'Longitude','Latitude']]
clean_us_starbucks_df
```

```python
# reset index
clean_us_starbucks_df.reset_index(drop=True, inplace=True)

clean_us_starbucks_df
```

```python
#Rename columns
clean_us_starbucks_df = clean_us_starbucks_df.rename(columns={"Store Number":"storenumber",
                                          "Store Name":"storename",
                                          "Street Address":"streetaddress",
                                          "City":"city",
                                          "State/Province":"state",
                                          "Country":"country",
                                          "five_digit_zip": "zipcode",
                                          "Phone Number":"phonenumber",
                                          "Longitude":"longitude",
                                          "Latitude":"latitude"})
clean_us_starbucks_df
```

```python
# LOADING: Starbucks file

# from config_Starbucks import username, password
# rds_connection_string = f"{username}:{password}@localhost:5432/starbucks_income_db"
# engine = create_engine(f'postgresql://{rds_connection_string}')

rds_connection_string = "postgres:***@localhost:5432/starbucks_test1"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Check for existing tables
engine.table_names()
```

```python
#Use pandas to load csv converted DataFrame into database
clean_us_starbucks_df.to_sql(name='starbucks_data_test', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from starbucks_data_test', con=engine) 
```

# ****IRS INCOME FILE*** 

```python

# reviewing Income file

csv_file = "resources/IRSIncomeByZipCode.csv"
income_df = pd.read_csv(csv_file)
income_df
```

```python
#Drop any NaN values that may exist (?)

income_df['ZIPCODE'] = income_df['ZIPCODE'].dropna()
income_df['ZIPCODE']
```

```python
# convert ZIPCODE Column "dtype" to a string 
income_df['ZIPCODE'] = income_df['ZIPCODE'].apply(str)
```

```python
# add  0 to zip code to 4-digit Postcode

income_df['five_digit_zip'] = income_df['ZIPCODE'].str.zfill(5)
income_df['five_digit_zip']
```

```python
income_df
```

```python
# verify 4-digit zipcode was converted to 5-digits
# ZIPCODE = 3280

income_df_test = income_df.loc[income_df['ZIPCODE'] == '3280']  
income_df_test
```

```python
# select column names from df
# income_df.columns 

clean_income_df = income_df[['STATE', 'five_digit_zip', 'Adjusted gross income (AGI)',
       'Avg AGI', 'Total income amount', 'Avg total income', 'Taxable income amount', 'Avg taxable income',]]
clean_income_df
```

```python
# reset index
clean_income_df.reset_index(drop=True, inplace=True)

clean_income_df

```

```python
#Rename columns
# clean_income_df.columns



clean_income_df = clean_income_df.rename(columns={"STATE" : "state",
                                                  "five_digit_zip": "zipcode",
                                                  "Adjusted gross income (AGI)": "adjusted_gross_income",
                                                  "Avg AGI": "avg_agi",
                                                  "Total income amount": "total_income",
                                                  "Avg total income": "avg_total_income", 
                                                  "Taxable income amount":"taxable_income_amount", 
                                                  "Avg taxable income":"avg_taxable_income"})


clean_income_df 
```
