```python

import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
csv_file = "resources/directory.csv"
starbucks_df = pd.read_csv(csv_file)
starbucks_df
```

```python
new_starbucks_df = starbucks_df.loc[starbucks_df["Country"] == "US"]
```

```python
new_starbucks_df
```

```python
new_starbucks_df['five_digit_zip'] = np.where(new_starbucks_df['Postcode'].str.len() == 9, 
                                     new_starbucks_df['Postcode'].str[:5],
                                     new_starbucks_df['Postcode'])
new_starbucks_df['five_digit_zip']
print(new_starbucks_df)
```

```python

new_starbucks_df['five_digit_zip'] = np.where(new_starbucks_df['five_digit_zip'].str.len() == 8, 
                                     new_starbucks_df['five_digit_zip'].str[:4],
                                     new_starbucks_df['five_digit_zip'])
print(new_starbucks_df)
```

```python
new_starbucks_df['five_digit_zip'] = new_starbucks_df['five_digit_zip'].str.zfill(5)
 
new_starbucks_df['five_digit_zip']
```

```python

clean_starbucks_df = new_starbucks_df[['Store Name','Street Address','City','State/Province','Country', 'five_digit_zip', 'Phone Number',
                    'Longitude','Latitude']]
clean_starbucks_df
```
