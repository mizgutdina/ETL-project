```python
US_df = df.loc[df['Country'] == 'US']
US_df
```

```python
US_df['Postcode'] = US_df['Postcode'].dropna()
US_df
```

```python
#Keep 5 digit Postalcode
#Condition: IF Postalcode > 5 digit, THEN chop last 4 digits, ELSE leave it alone
US_df['five_digit_zip'] = np.where(US_df['Postcode'].str.len() == 9, 
                                     US_df['Postcode'].str[:5],
                                     US_df['Postcode'])
print(US_df)
```

```python
US_df['five_digit_zip'] = np.where(US_df['five_digit_zip'].str.len() == 8, 
                                     US_df['five_digit_zip'].str[:4],
                                     US_df['five_digit_zip'])
print(US_df)
```
